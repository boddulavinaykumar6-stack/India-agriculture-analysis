###### Unique count and values of Crops, States and Districts

```python
unique_crop_list = agri_df['Crop'].unique()

print(f"Total no of unique crops - {len(unique_crop_list)}")
print(f"We have following unique crops in the dataset- \n{unique_crop_list}")
```

```python
unique_state_list = agri_df['State'].unique()

print(f"Total no of unique state - {len(unique_state_list)}")
print(f"We have following unique state in the dataset- \n{unique_state_list}")
```

```python
unique_district_list = agri_df['District'].unique()

print(f"Total no of unique district - {len(unique_district_list)}")
print(f"We have following unique district in the dataset- \n{unique_district_list}")
```

```python
agri_df['Production Units'].unique()
```

```python
agri_df[agri_df['Production Units'] == 'Nuts']['Crop'].unique()
```

```python
def unit_standardization(df):
    if df['Production Units'] == 'Nuts':
        return np.round(df['Production']/50,1)
    elif df['Production Units'] == 'Tonnes':
        return df['Production']
    else:
        return np.round(df['Production']/4.5929637955183,1)
```

```python
agri_df['New Production'] = agri_df.apply(unit_standardization,axis=1)
```

```python
agri_df.drop(columns=['Production','Production Units'],inplace=True)
```

```python
agri_df.head(3)
```

```python
#Which are the top 5 most common crops in India

agri_df['Crop'].value_counts().head()
```

```python
#Get statewise total production of crops
total_production_list = []
for state in unique_state_list:
    total_production_list.append(agri_df.loc[agri_df['State'] == state, 'New Production'].sum())
```

```python
crop_production_df = pd.DataFrame({
    'State' : unique_state_list,
    'Total Crop Production' : total_production_list
})
crop_production_df.sort_values('Total Crop Production',ascending=False).head()
```

### Highest producer of Rice since 1997

```python
agri_rice_df = agri_df[agri_df['Crop'] == 'Rice']
unique_rice_growing_states = agri_rice_df['State'].unique()

total_rice_production = []
for state in unique_rice_growing_states:
    total_rice_production.append(np.round(agri_rice_df.loc[agri_rice_df['State'] == state, 'New Production'].sum(),2))
    
```

```python
rice_production_df = pd.DataFrame({
    'State' : unique_rice_growing_states,
    'Total Rice Production' : total_rice_production
})
```

```python
rice_production_df.sort_values('Total Rice Production',ascending=False).head()
```

###### Best year for Agriculture

```python
unique_year_list = agri_df['Year'].unique()
```

```python
unique_year_list
```

```python
yearly_production_list = []
for year in unique_year_list:
    yearly_production_list.append(agri_df.loc[agri_df['Year']==year, 'New Production'].sum())
```

```python
yearly_production_df = pd.DataFrame(
{
    'year' : unique_year_list,
    'total crop production' : yearly_production_list
})
yearly_production_df.sort_values('total crop production', ascending=False).head()
```
